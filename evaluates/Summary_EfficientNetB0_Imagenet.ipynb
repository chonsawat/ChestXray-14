```python
import pandas as pd
import tensorflow as tf
```

```python
MODEL_NAME = "EfficientNetB0" # DenseNet121 / EfficientNetB0 / Resnet50
WEIGHT_OPTION = "Imagenet" # None / Imagenet
LOSS_FUCNCTION = "cross_entropy"
EXPERIMENT_NAME = f"{MODEL_NAME}_{WEIGHT_OPTION}"
EXPERIMENT_PATH = f"results/evaluate/{LOSS_FUCNCTION}/{EXPERIMENT_NAME}"
```

```python
f1_scores = []
for input_path in tf.io.gfile.glob(f"{EXPERIMENT_PATH}/*/f1_scores.csv"):
    df = pd.read_csv(input_path)
    f1_scores.append(df)
    
df_f1_scores = pd.concat(f1_scores)
df_f1_scores.to_csv(f"{EXPERIMENT_PATH}/f1_scores.csv", index=False)
df_f1_scores
```

```python
precision_scores = []
for input_path in tf.io.gfile.glob(f"{EXPERIMENT_PATH}/*/precision.csv"):
    df = pd.read_csv(input_path)
    precision_scores.append(df)
    
df_precision_scores = pd.concat(precision_scores)
df_precision_scores.to_csv(f"{EXPERIMENT_PATH}/precision.csv", index=False)
df_precision_scores
```

```python
recall_scores = []
for input_path in tf.io.gfile.glob(f"{EXPERIMENT_PATH}/*/recall.csv"):
    df = pd.read_csv(input_path)
    recall_scores.append(df)
    
df_recall_scores = pd.concat(recall_scores)
df_recall_scores.to_csv(f"{EXPERIMENT_PATH}/recall.csv", index=False)
df_recall_scores
```
