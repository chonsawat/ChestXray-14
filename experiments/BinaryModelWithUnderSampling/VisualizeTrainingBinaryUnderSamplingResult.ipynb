````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from pathlib import Path
VISUALIZE_OUTPUT_PATH = "results/visualize"
Path(VISUALIZE_OUTPUT_PATH).mkdir(parents=True, exist_ok=True)
````

````python
sns.set_theme()
````

## Utils

````python
def plot_traning_history(path, filename):
    df = pd.read_csv(path)
    fig = plt.figure(figsize=(12, 8))
    plt.subplots_adjust(hspace=0.4)

    plt.subplot(211)
    plt.ylim((-0.05, 0.55))
    sns.lineplot(data=df, x=df.columns[0], y=df.columns[2], label="loss")
    sns.lineplot(data=df, x=df.columns[0], y=df.columns[4], label="val_loss")

    best = df[df.columns[4]].argmin()
    sns.scatterplot(x=[df[df.columns[0]][best]], y=[df[df.columns[4]][best]], marker="X", s=200, color="red")
    plt.title("Training Loss")

    plt.subplot(212)
    plt.ylim((0.45, 1.05))
    sns.lineplot(data=df, x=df.columns[0], y=df.columns[1], label="auc")
    sns.lineplot(data=df, x=df.columns[0], y=df.columns[3], label="val_auc")

    best = df[df.columns[4]].argmin()
    sns.scatterplot(x=[df[df.columns[0]][best]], y=[df[df.columns[3]][best]], marker="X", s=200, color="red")
    plt.title("Training AUC");

    fig.savefig(f"{VISUALIZE_OUTPUT_PATH}/{filename}.png", format="png")
````
