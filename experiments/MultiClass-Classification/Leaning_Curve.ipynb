```python
EXPERIMENT_NAME = "multiclass_classification_with_cross_entropy_loss_2"
MODEL_NAME = "EfficientNetB0"
WEIGHT = "None"
FOLD_NUM = 1
```

```python
history = pd.read_csv(f"{RESULT_PATH}/history/{EXPERIMENT_NAME}/{MODEL_NAME}_{WEIGHT}_FOLD_{FOLD_NUM}/history.csv")
history
```

```python
plt.plot(history.loss, label="Training");
plt.plot(history.val_loss, label="Validation");
plt.legend();
```

```python
def parse_f1(value):
    
    return np.mean(eval(eval(value)))
```

```python
plt.plot(history.f1_score.apply(parse_f1), label="Training");
plt.plot(history.val_f1_score.apply(parse_f1), label="Validation");
plt.legend();
```
