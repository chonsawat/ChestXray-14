```python
from utils import *
import tensorflow as tf
import matplotlib.pyplot as plt
input_path = "dataset/ChestXray NIH"
```

```python
train_filenames = tf.io.gfile.glob(f'{input_path}/data/224x224/train/*.tfrec')
train_dataset = get_dataset(train_filenames, shuffled=False, repeated=False, augmented=False)
```

```python
plt.figure(figsize=(8,8))
for element in train_dataset.batch(1):
    plt.subplot(221)
    plt.imshow(element[0][0][0])
    plt.axis(False)
    
    plt.subplot(222)
    plt.imshow(element[0][0][1])
    plt.axis(False)
    
    plt.subplot(223)
    plt.imshow(element[0][0][2])
    plt.axis(False)
    
    plt.subplot(224)
    plt.imshow(element[0][0][3])
    plt.axis(False)
    
    break
```
