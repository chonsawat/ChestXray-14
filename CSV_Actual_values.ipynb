```python
import numpy as np
import pandas as pd
import tensorflow as tf
from modules.dataset import LABELS, Dataset
```

```python
def get_test_dataset_5_fold():
    dataset = Dataset()
    _, test_dataset_fold_1 = dataset.get_kfold(fold_number=1, sample=False)
    _, test_dataset_fold_2 = dataset.get_kfold(fold_number=2, sample=False)
    _, test_dataset_fold_3 = dataset.get_kfold(fold_number=3, sample=False)
    _, test_dataset_fold_4 = dataset.get_kfold(fold_number=4, sample=False)
    _, test_dataset_fold_5 = dataset.get_kfold(fold_number=5, sample=False)
    return (
        test_dataset_fold_1,
        test_dataset_fold_2,
        test_dataset_fold_3,
        test_dataset_fold_4,
        test_dataset_fold_5
    )
```

```python
def get_y_true(data):
    y_true=[]
    for X,y in data:
        for label in y:
            y_true.append(label)
    y_true = tf.Variable(y_true)
    return y_true.numpy().tolist()
```

```python
path = "/home/jovyan/ChestXray-14/results/actual_values"
dataset = get_test_dataset_5_fold()
```

```python
N = 5

data_dict = dict()
for i in range(0, N):
    data_dict[f"Fold {i+1}"] = get_y_true(dataset[i])
```

```python
df = pd.DataFrame(data_dict)
df
```

```python
df.to_csv(f"{path}/actual.csv", index=False)
```
